# **Einleitung**

## **Stroke Prediction Dataset**

Der Stroke Prediction Dataset ist ein umfangreicher Datensatz, der Informationen über verschiedene Kategorien von Merkmalen von Personen enthält. Diese Kategorien decken soziale, gesundheitliche und andere relevante Aspekte ab und dienen zur Vorhersage von Schlaganfällen. 

### **Beschreibung der Merkmale**

- **Soziale Merkmale**:
  - **gender**: Das Geschlecht der Person  (Male (männlich), Female (weiblich) und Other (Anderes Geschlecht)).
  - **age**: Das Alter der Person in Jahren.
  - **ever_married**: Antwort auf die Frage, ob die Person bereits verheiratet war (Yes/No)
  - **work_type**: Der Typ der Beschäftigung der Person (Private (privat), Self-employed (selbstständig), Govt_job (Regierungsangestellter), children (Kind), Never_worked (nie gearbeitet)).
  - **Residence_type**: Der Wohnort der Person, entweder "Rural" (ländliche Umgebung) oder "Urban" (städtische Umgebung).


- **Gesundheitliche Merkmale**:
  - **hypertension**: Ein binärer Wert (1, wenn die Person an Bluthochdruck leidet, 0 sonst).
  - **heart_disease**: Ein binärer Wert (1, wenn die Person an einer Herzerkrankung leidet, 0 sonst).
  - **avg_glucose_level**: Der durchschnittliche Glukosespiegel im Blut der Person.
  - **bmi**: Der Body-Mass-Index (BMI) der Person.
  - **smoking_status**: Der Raucherstatus der Person (z. B. never smoked (nie geraucht), formerly smoked (ehemaliger Raucher), smokes (aktueller Raucher), Unknown (unbekannte Antwort)).
  - **stroke**: Binäre Zielvariable (1, wenn die Person einen Schlaganfall erlitt, 0 sonst)

Diese Merkmale bieten Einblicke in den sozialen Hintergrund, den allgemeinen Gesundheitszustand und das Verhalten der Personen, die im Datensatz enthalten sind. Indem verschiedene Kategorien berücksichtigt werden, können Modelle zur Schlaganfallvorhersage fundiertere Entscheidungen treffen und ein besseres Verständnis für die zugrunde liegenden Risikofaktoren entwickeln.

## **3.1 Merkmale**

Bevor wir mit der explorativen Analyse beginnen können, ist es wichtig, die Datentypen der einzelnen Merkmale (Features) in unserem Datensatz zu überprüfen. Dies ist entscheidend, um die richtigen Darstellungsweisen für die explorative Analyse und Visualisierung auszuwählen. Merkmale lassen sich üblicherweise in drei Datentypen kategorisieren:
1. **Nominal:** Merkmale, die keine natürliche Reihenfolge haben. Ein Beispiel für ein solches Merkmal ist das Betriebssystem (Windows, MacOS, Linux).
2. **Ordinal:** Merkmale mit einer geordneten Reihenfolge. Ein Beispiel hierfür ist die T-Shirt-Größe (S, M, L).
3. **Quantitativ:** Merkmale, die numerische Messungen oder Größen repräsentieren. Quantitative Merkmale können weiterhin in diskrete und kontinuierliche Merkmale unterteilt werden. Diskrete Merkmale haben nur wenige mögliche Werte. Ein Beispiel für ein diskretes Merkmal ist die Anzahl der Kinder in einer Familie, da hier nur ganze Werte möglich sind. Ein Beispiel für ein kontinuierliches Merkmal wäre die Temperatur, da sie unendlich viele mögliche Werte annehmen kann.

Indem wir die Datentypen identifizieren, können wir die richtigen statistischen Analysen und Visualisierungstechniken auswählen, um die Merkmale angemessen zu untersuchen. Die Merkmale können auch Einfluss auf die Wahl des Modells haben, welches wir auf die Daten anwenden.

In Pandas ist es möglich, den Datentyp einer Spalte auszugeben. Der Pandas-Datentyp kann beispielsweise ein Integer, Boolean oder Float sein. Es ist jedoch wichtig zu beachten, dass der Pandas-Datentyp nicht unbedingt mit dem des zugrunde liegenden Merkmals übereinstimmt. Dennoch kann der Pandas-Datentyp als Orientierung und Ausgangspunkt dienen, um den Datentyp des Merkmals zu ermitteln. Bei der Bestimmung der Datentypen der Merkmale ist es auch ratsam, vorhandene Informationen und Dokumentationen zu den Merkmalen zu berücksichtigen. Oftmals liefern Datenquellen oder Datenbeschreibungen Hinweise darauf, welchen Datentyp ein bestimmtes Merkmal haben sollte. Dies kann bei der korrekten Zuordnung der Merkmale in die Kategorien nominales, ordinäres oder quantitatives Merkmal helfen.

Der vorliegende Datensatz über Schlaganfälle enthält insgesamt zehn Merkmale. Durch Ausführen der Pandas-Methode "info()" auf dem Trainingsdatensatz können die Datentypen der Merkmale in der Spalte "Dtype" angezeigt werden. Dabei zeigt sich, dass es insgesamt drei Float-, zwei Integer- und fünf Object-Datentypen gibt. Der Object-Datentyp ist äquivalent zu einem Python Object und stellt in unserem Anwendungsfall einen String-Datentyp dar.

Um nun den Merkmalsdatentyp genauer zu bestimmen, kann neben dem "DType" die Methode "value_counts()" von Pandas verwendet werden. Diese Methode liefert alle vorhandenen Werte in einem Merkmal und zeigt, wie häufig sie jeweils vorkommen. Durch die Betrachtung der eindeutigen Werte und ihrer Häufigkeiten kann auf den Merkmalsdatentyp geschlossen werden.

## **3.3 Anpassung des Trainingsdatensatzes für die explorative Analyse**

Um die explorative Analyse zu vereinfachen wird der Trainingsdatensatz angepasst. Zuerst werden die Trainingsdatensätze der Merkmale(X_train) und der Zielvariablen(Y_train) nach dem ursprünglichen Index der Einträge zusammengeführt. Außerdem werden die Spalten der Merkmale "hypertension" und "heart_disease" verändert. Diese Merkmale sind im Ursprungsdatensatz numerisch angegeben, wobei eine 1 semantisch für "Yes" und eine 0 semantisch für "No" steht. Da diese Merkmale nominal sind, ist es sinnvoll die numerischen Werte auf die äquivalenten String-Repräsentationen, "Yes" und "No", zu ändern. 

````python
X_train.head()
````

````python
# Merge X_train und Y_train um ein DF mit Trainings Features und Zielvariablen zu haben
merged_train = X_train.merge(Y_train, left_index=True, right_index=True)

#Replace 1 with Yes and 0 with No
merged_train["hypertension"] = merged_train["hypertension"].replace({0: "No", 1: "Yes"})
merged_train["heart_disease"] = merged_train["heart_disease"].replace({0: "No", 1: "Yes"})

merged_train.head()
````

## **3.4 Datensatz im Überblick - Visualisierungen**

Im Folgenden werden Visualisierungen erstellt, um einen umfassenden Überblick über den Datensatz und seine Merkmale, besonders im Bezug auf die Zielvariable, zu erhalten.


Um Merkmale zu visualisieren gibt es verschiedene Techniken und Richtlinien. Um **zwei quantitative Merkmale** zu visualisieren eignet sich in der Regel ein Streudiagramm (scatter plot). Jeder Punkt stellt dabei den Zusammenhang der Merkmale dar. Die Wertekombinationen der Merkmale sind als genaue Position im Streudiagramm abzulesen. In dieser Art des Diagramms wird besonders auf lineare oder einfache nicht-linear Zusammenhänge geschaut.  Diese Zusammenhänge stellen die stärke des Zusammenhangs der beiden Merkmale dar. 

Bei **einem qualitativen und einem quantitativen Merkmal** wird das qualitative Merkmal oftmals gruppiert und das qualitative Merkmal innerhalb dieser Gruppen betrachtet. Die Darstellung dieser Merkmale kann durch Dichtekurven (density curves) oder nebeneinander dargestellte Boxplots geschehen. Durch die Dichtekurven lässt sich die Verteilung der einzelnen Merkmalsgruppen darstellen. Hier sollte besonders auf die Art der Verteilung und die Ausbreitung geschaut werden. Dichtekurven implizieren, dass die gesamte Flächer der Balken 1 ergibt. Der Boxplot stellt die Verteilung der Merkmalsgruppen in einer simpeleren Form dar. Das besondere an Boxplots ist, dass sie mehrere statistische Werte in einer Darstellung abbilden. Die Box geht dabei von dem 25 Perzentil bis zum 75 Perzentil der Daten. Der Strich in der Box zeigt den Median der Werte des Merkmals. Außerdem werden durch Punkte unterhalb und oberhalb des Boxplots, Ausreißer angezeigt. Boxplots können nicht die selben Informationen über das Aussehen der Verteilung wie die Dichtekurve geben, haben aber den Vorteil, dass mehrere Aussagen über Ausreißer und die Symmetrie der Daten getroffen werden können.

Bei der Betrachtung **zweier qualitativer Merkmale** wird wieder gerne ein Merkmal in Untergruppen unterteilt. Das heißt, dass ein Merkmal konstant gehalten wird und die Verteilung des anderen Merkmals dargestellt wird. Die Visualisierung dieser Merkmale kann durch ein Balkendiagramm (bar chart) oder ein Liniendiagramm (line chart) stattfinden. 

Sollen mehr **multivariate Merkmale** dargestellt werden, wollen wir diese wieder über Untergruppen darstellen. Eine solche Visualisierung kann über ein Streudiagramm stattfinden, wobei verschiedenen Kategorien durch unterschiedliche Symbole dargestellt werden. Je größer die Anzahl an Merkmalen ist, die betrachtet werden sollen, desto schwieriger wird die Visualisierung jener. Ein weiterer Ansatz kann sein die Darstellung auf mehrere Subdarstellungen aufzuteilen.

Weiterhin ist wichtig zu beachten, dass die Visualisierung von qualitativen kategorischen Merkmalen mit steigender Anzahl der Kategorein an Schwierigkeit gewinnt. Umso mehr Kategorien es gibt, desto größer ist die Anzahl der Untergruppen und auch der Darstellungen, die für ein Merkmal gezeichnet werden müssen.

# **4. Cleaning und Skalierung der Merkmale und Transformation**

## **4.1 Data cleanup**

Da die meisten Algorithmen des maschinellen Lernens nicht gut mit fehlenden Werten umgehen können, ist es sinnvoll, diese zu behandeln. Grundsätzlich gibt es drei Methoden, um dieses Problem zu lösen. Die erste Methode ist es das gesamte Merkmal zu ignorieren. Die zweite Methode besteht darin, das Attribut aus dem Datensatz zu entfernen und die letzte Methode besteht darin, die fehlenden Werte durch andere Werte zu ersetzen. Die Ersetzungsmethode wird auch "Imputation" genannt. In unserem Datensatz haben wir uns für die Imputationsmethode entschieden. Bei dieser Methode werden bei numerischen Merkmalen alle fehlenden Werte z.B. durch den Median ersetzt. Da diese Methode nur auf numerische Werte angewendet werden kann, müssen Textattribute in numerische Attribute umgewandelt werden. Eine einfache Umwandlung von kategorialen Merkmalen in numerische Werte ist jedoch nicht immer sinnvoll, da maschinelle Lernalgorithmen oft davon ausgehen, dass eine näher liegende Zahl ähnlicher ist als eine weiter entfernt liegende Zahl. Diese Annahme würde jedoch beispielsweise nicht zutreffen, wenn wir für das Merkmal Art der Beschäftigung den Wert 1 für privat, den Wert 2 für selbständig und den Wert 3 für öffentlich definieren würden. In diesem Fall würde der Algorithmus annehmen, dass eine Person, die privat beschäftigt ist, einer Person, die selbständig beschäftigt ist, ähnlicher ist als einer Person, die öffentlich beschäftigt ist. Um diese Verzerrung zu korrigieren, verwenden wir den OneHotEncoder von Sklearn. Der OneHotEncoder erstellt eine binäre Variable pro Kategorie. [1]

## **4.2 Skalierung und Transformation**

Da maschinelle Lernalgorithmen bei numerischen Merkmalen mit unterschiedlicher Skalierung schlechter abschneiden, ist es wichtig, die Skalierung anzupassen. In der Praxis gibt es zwei Möglichkeiten, die Skalierung zu vereinheitlichen. Zum einen gibt es die Min-Max-Methode oder auch Normalisierung und zum anderen die Standardisierung.

- Bei der Normalisierung werden die Werte so umskaliert, dass sie zwischen 0 und 1 liegen. Die Skalierung erfolgt durch Subtraktion des minimalen Wertes und Division durch die Differenz zwischen dem minimalen und dem maximalen Wert. Sklearn stellt für die Normalisierung den MinMaxScaler zur Verfügung.

- Bei der Standardisierung wird zunächst der Durchschnittswert subtrahiert und das Ergebnis durch die Standardabweichung geteilt. Die standardisierten Werte haben somit eine Standardabweichung von 1. Im Gegensatz zur Normalisierung werden die Werte bei der Standardisierung nicht in einen bestimmten Bereich eingeordnet. Der Vorteil der Standardisierung ist, dass sie weniger von Ausreißern beeinflusst wird als die Normalisierung. Um die Standarisierung zu verwenden, stellt Sklearn den StandardScaler zur Verfügung.

Bei der Verwendung von Skalierern ist zu beachten, dass diese wie die Modelle nur an den Trainingsdaten angepasst werden, um die Modellparameter zu bestimmen. Sobald ein trainierter Skalierer zur Verfügung steht, kann dieser auch auf die Testdaten angewendet werden. In unserem Fall haben wir uns für die Standardisierung der Daten entschieden und den StandardScaler von Sklearn verwendet.

Da die Skalierung dafür sorgt, dass die Merkmalswerte in einem kleinen Bereich liegen, ist es sinnvoll, einige Merkmale zu transformieren. Mit dieser Methode sollen, "heavy tails" vermeiden und symmetrische Verteilungen zu erzeugt werden. Um dies zu realisieren, wird häufig die Quadratwurzel oder der Logarithmus von numerischen Merkmalen genommen. Um Merkmale mit einer bimodalen Verteilung zu transformieren, kann die "Bucketizing"-Methode zu benutzen. Dabei wird die Verteilung in Töpfe aufgeteilt und jeder Merkmalswert durch eine seinem Topf entsprechende ID ersetzt. Die IDs müssen dann erneut "One Hot" kodiert werden, um die numerische Konsistenz der Merkmale beizubehalten.


[1]

## **4.3 Pipeline für die Transformation und Skalierung der Daten**

Um Datenbereinigung, Skalierung und Transformation an einem Datensatz durchzuführen, haben wir diese Schritte in einer Pipeline zusammengefasst. Zunächst werden getrennte Pipelines für kategoriale (categorical_transformer) und numerische Merkmale (numerical_transformer) erstellt. Die kategoriale Pipeline behandelt fehlende Werte, indem sie diese mit den häufigsten Wert einsetzt, und wendet dann das sogenannte "One-Hot Encoding" an, um kategoriale Merkmale in binäre Merkmale umzuwandeln; die numerische Pipeline behandelt fehlende Werte, indem sie den Medianwert einsetzt, und wendet dann die Quadratwurzelfunktion auf die Merkmale an. Der nächste Schritt ist der "ColumnTransformer", der die kategoriale und die numerische Pipeline auf die entsprechenden Spalten des Datensatzes anwendet. Die Spalten 'gender', 'hypertension', 'heart_disease', 'ever_married', 'work_type', 'residence_type', 'smoking_status' werden mit der kategorialen Pipeline und die Spalte 'bmi' mit der numerischen Pipeline transformiert. Ein weiterer Schritt ist der "KBinsDiscretizer", der die Merkmale 'age' und 'avg_glucose_level' in diskrete Bins einteilt. In diesem Fall werden 5 Bins mit gleicher Breite verwendet und die Bin-Labels werden als "onehot-dense" kodiert. Schließlich wird die Gesamtpipeline erstellt, indem die vorherigen Schritte in einer Hauptpipeline kombiniert werden. Diese wird dann auf den Trainingsdatensatz (X_train) angewendet, um die Transformationen durchzuführen, und anschließend auf den Testdatensatz (X_test), um die Konsistenz der Merkmalswerte zu gewährleisten.

Im Folgenden ist ein Auszug aus dem Trainingsdatensatz nach Durchlaufen der Pipeline dargestellt:

````python
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(handle_unknown='ignore'))
])

numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('sqrt_transformer', FunctionTransformer(np.sqrt, feature_names_out="one-to-one")),
    ('scaler', StandardScaler())
])


preprocessor = ColumnTransformer(transformers=[
    ('categorical', categorical_transformer, ['gender', 'hypertension', 'heart_disease', 'ever_married', 'work_type', 'Residence_type', 'smoking_status']),
    ('numerical', numerical_transformer, ['bmi']) 
])

preprocessor = ColumnTransformer(transformers=[
    ('preprocessor', preprocessor, ['gender', 'hypertension', 'heart_disease', 'ever_married', 'work_type', 'Residence_type', 'smoking_status', 'bmi']),
    ('bucketizer', KBinsDiscretizer(n_bins=5, strategy='uniform', encode='onehot-dense'), ['age', 'avg_glucose_level'])
])

pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor)
])

X_train_transformed = pipeline.fit_transform(X_train)
X_test_transformed = pipeline.transform(X_test)

X_train = pd.DataFrame(X_train_transformed, columns=pipeline.get_feature_names_out())
X_test = pd.DataFrame(X_test_transformed, columns=pipeline.get_feature_names_out())

X_train.head(5)
````

## **5.1 Resampling der Daten**

Ein Hauptproblem des Schlaganfalldatensatzes für das Training der Modelle ist die Unausgewogenheit der Klassen. Wie die explorative Analyse gezeigt hat, ist die Wahrscheinlichkeit, dass kein Schlaganfall vorliegt (ca. 95 %), deutlich höher als die Wahrscheinlichkeit, dass ein Schlaganfall vorliegt. Um dieses Problem zu beheben, können die Verhältnisse der Klassen im Trainingsdatensatz verändert werden. Dazu gibt es zwei wichtige Techniken:

- **(Random) Oversampling:** Erhöhung der Stichproben im Datensatz. Es werden Daten der unterrepräsentierten Klasse hinzugefügt.
- **(Random) Undersampling:** Reduktion der Stichproben im Trainingsdatensatz. Daten der überrepräsentierten Klasse werden ohne sie zu ersetzen entfernt.

Wir haben uns für die Methode des Oversamplings in diesem Datensatz entschieden, da die Gesamtmenge der Daten klein ist (ca. 4000). Um Overfitting zu vermeiden, haben wir uns entschieden, das Oversampling nur auf den Trainingsdatensatz anzuwenden. Dennoch erhöht die steigende Anzahl der replizierten Stichproben im Trainingsdatensatz das Risiko des Overfittings. Außerdem geht mit dem Oversampling einher, dass sich die Trainingszeit der Modelle durch den Zugewinn an Daten erhöht.

````python
def addMLAlgoToRanking(algoName):
    mlAlgoRanking[algoName] = {"f1-score" : f1_score(Y_test, y_pred), "Best parameters" : best_params, "Recall": recall_score(Y_test, y_pred), "Specificity": specificity_score(Y_test, y_pred), "Precision": precision_score(Y_test, y_pred)}


def rankMLAlgos():
    df = pd.DataFrame.from_dict(mlAlgoRanking, orient='index')
    df_sorted = df.sort_values(by='f1-score', ascending=False)
    df_sorted['f1-score'] = df_sorted['f1-score'].apply(lambda x: round(x, 4))
    recalls = df_sorted["Recall"].to_list()
    specificities = df_sorted["Specificity"].to_list()
    precisions = df_sorted["Precision"].to_list()
    labels = df_sorted.index.to_list()

    colors = ['red', 'green', 'blue', 'orange', 'purple', 'yellow', 'black']  # Liste der Farben

    fig, ax = plt.subplots()
    for i in range(len(recalls)):
        ax.scatter(recalls[i], precisions[i], color=colors[i])

    ax.set_xlabel("Recall")
    ax.set_ylabel('Precisions')
    ax.set_title('Plot of Recalls against Precisions')

    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.grid(True)

    handles = []
    for i, label in enumerate(labels):
        handles.append(ax.scatter([], [], color=colors[i], label=label))
    ax.legend(handles, labels, loc='lower right', frameon=False, bbox_to_anchor=(1.5, 0.5))

    plt.savefig("Comparison of all Models.svg", format="svg")
    plt.show()



def plotConfusionMatrix(Y_test, y_pred):
    cm = confusion_matrix(Y_test, y_pred)
    color_palette = ListedColormap(['#063d38', '#34685b', '#a3caab', '#f16800'])
    sns.heatmap(cm, annot=True, fmt="d", cmap=color_palette)
    plt.xlabel("Vorhergesagt")
    plt.ylabel("Tatsächlich")
    plt.title("Wahrheits Matrix")
    plt.savefig("Konfusionsmatrix.svg", format="svg")
    plt.show()

def calculateMetrics(Y_test, y_pred):
    accuracy = accuracy_score(Y_test, y_pred)
    precision = precision_score(Y_test, y_pred)
    recall = recall_score(Y_test, y_pred)
    specificity = specificity_score(Y_test, y_pred)
    f1 = f1_score(Y_test, y_pred)

    metrics_data = {'Metrik:': ['Accuracy:', 'Precision:', 'Sensitivity (Recall):', 'Specificity:', 'F1-Score:'],
            'Werte:': [accuracy, precision, recall, specificity, f1]}
    metrics_df = pd.DataFrame(metrics_data)
    display(metrics_df)


def best_params_table(best_params):
    param_list = []
    for key, value in best_params.items():
        param_list.append({'Parameter:': key, 'Wert:': value})

    params_df = pd.DataFrame(param_list)
    display(params_df)


def modelEvaluation(Y_test, y_pred, best_params):
    calculateMetrics(Y_test, y_pred)
    best_params_table(best_params)
    plotConfusionMatrix(Y_test, y_pred)

def printCrossValidationF1(grid_search):
    f1_scores_validation = grid_search.cv_results_['mean_test_score'].tolist()
    for i, f1 in enumerate(f1_scores_validation):
        print(f"F1-Score für Validierungssatz {i+1}: {f1}")

def plotROC(best_model, model):
    y_scores = best_model.predict_proba(X_test)[:, 1]
    
    fpr, tpr, thresholds = roc_curve(Y_test, y_scores)
    roc_auc = auc(fpr, tpr)
    plt.figure()
    plt.plot(fpr, tpr, color="#34685b", lw=2, label='AUC (area = %0.2f)' % roc_auc)
    plt.plot([0, 1], [0, 1],  color="#f16800", lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('Falsch-Positiv-Rate (fall-out / 1-Sensitivität)')
    plt.ylabel('Richtig-Positiv-Rate (Sensitivität / Recall)')
    plt.title(f"ROC Kurve - {model} - Test Daten")
    plt.legend(loc="lower right")
    plt.grid(color='gray', linestyle='--', linewidth=0.5)
    
    ax = plt.gca()
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.yaxis.set_ticks_position('none')
    ax.xaxis.set_ticks_position('none')

    plt.savefig("ROC-Curve.svg",format="svg")
    plt.show()

def gridSearchAndCV(model):
    grid_search = GridSearchCV(model, param_grid, cv=5, scoring='recall')
    grid_search.fit(X_train, Y_train)

    best_model = grid_search.best_estimator_
    best_params = grid_search.best_params_
    best_model.fit(X_train, Y_train)

    y_pred = best_model.predict(X_test)
    return y_pred, best_model, best_params
````

### **Kreuzvalidierung**

Die Kreuzvalidierung dient dazu, die Leistungsfähigkeit eines maschinellen Lernmodells genauer zu beurteilen, indem ein Modell mehrmals auf verschiedenen Trainingsdatensätzen trainiert wird. Die Daten werden in k gleich große, disjunkte Teilmengen, so genannte Folds, aufgeteilt. Anschließend werden k Modelle trainiert, wobei für jedes Modell eine Kreuzvalidierung durchgeführt wird. Das bedeutet, dass pro Durchlauf ein Fold als Testdatensatz dient und das Modell auf den restlichen Datensätzen trainiert wird. In den verbleibenden Durchläufen wird das Verfahren wiederholt, so dass insgesamt jeder Datensatz einmal zur Validierung verwendet wurde. Dadurch wird sichergestellt, dass sich jeder Datenpunkt einmal im Testdatensatz befindet und somit das Modell für alle Datenpunkte funktionieren muss um eine hohe Genauigkeit des Modells zu erreichen. Ebenso werden in jedem Durchlauf $(1-\frac{1}{k})\%$ zum Training benutzt, statt den üblichen $25\%$ für den einzigen Durchlauf ohne Kreuzalidierung. Jedoch müssen nunn auch k Modelle trainiert werden, was zu einem k-vielfachen Rechenaufwand sowie k-fachen Zeitaufwand führt. [4, Ch. 5 Evaluierung und Verbesserung von Modellen]

<!--[4] 5 Evaluierung Und Verbesserung Von Modellen
(Quelle: Müller, Andreas C./Sarah Guido (2017): Einführung in Machine Learning mit Python: Praxiswissen Data Science, O’Reilly.)-->

### **Grid Search**

Der zweite Schritt zur Verbesserung der Modellleistung ist die optimale Bestimmung der Modellparameter. Dies kann z.B. mit Hilfe der Gitter-Suche erfolgen. Dabei werden alle möglichen Kombinationen der zuvor definierten Parameter getestet, die dem Modell als Dictionary übergeben werden. Das Modell mit den optimalen Parametern hat dann die höchste durchschnittliche Güte der Kreuzvalidierungsdurchläufe.

Für das optimale Ergebnis werden die beiden geannten Methoden kombiniert. Dafür stellt ```sklearn``` die Klasse ```GridSearchCV``` zur Verfügung. In Folge dessen müssen anschließend insgesamt $\#Permutationen \ der\ Parameter\ \cdot \ k\ $-Modelle trainiert werden, was die Rechenzeit der Modelle noch einmal deutlich erhöht. Bei beispielsweise zwei Parametern mit jeweils drei Optionen und einer fünffachen Kreuzvalidierung landet man schon bei $3! \ \cdot \ 3!\ \cdot 5\ = 180$ Modellen, die trainiert werden müssen. ### **Konfusionsmatrix**

Konfusionsmatrizen sind 2 x 2 Matrizen zur Darstellung der Ergebnisse binärer Klassifikatoren. Die Spalten entsprechen den vorhergesagten Kategorien (positive oder negative Klasse), während die Zeilen die tatsächlichen Kategorien darstellen. Die Einträge in der Matrix geben die absolute Häufigkeit der Datenpunkte an, die den Kategorien in den Zeilen zugeordnet wurden. Die Einträge auf den Hauptdiagonalen oben links und oben rechts wurden richtig klassifiziert, während die anderen Einträge den falschen Kategorien zugeordnet wurden. Daher werden die Datenpunkte, die korrekt mit der positiven Klasse identifiziert wurden, als "true positive" bezeichnet, oft abgekürzt als TP für "true positive", und die Datenpunkte, die korrekt mit der negativen Klasse identifiziert wurden, als "true negative", oft abgekürzt als TN für "true negative". Einträge der negativen Klasse, die der positiven Klasse zugeordnet wurden, werden als FP ("false positive") bezeichnet und Einträge der positiven Klasse, die als negativ klassifiziert wurden, als FN ("false negative"). Diese Zusammenhänge sind auch in der folgenden beispielhaften Konfusionsmatrix dargestellt. [4, Ch. 5 Evaluierung und Verbesserung von Modellen]

### **Accuracy**
Die Genauigkeit (Accuracy) entspricht den Anteil aller korrekt klassifizierten Datenpunkte an der Gesamtmenge aller klassifizierten Datenpunkte. Entsprächend lässt sie sich anhand folgender Formel berechnen.

$Accuracy\ = \frac{TP+TN}{TP+TN+FP+FN}$

$Precision \ = \ \frac{TP}{TP+FP}$

### **Sensitivity (Recall)**
Die Sensitivität (Sensitivity oder Recall) entspricht den Anteil aller korrekt klassifizierten Datenpunkte der positiven Klasse an der Gesamtmenge der positiven Klasse unabhängig von ihrer Klassifizierung. Daraus lässt sich ableiten, dass die Anzahl der falsch negativen Datenpunkte minimiert werden soll. Dies spielt zum Beispiel eine Rolle beim Identifizieren kranker Personen. Die Sensitivität lässt sich über die folgende Formel bestimmen.

$Sensitivity\ = \frac{TP}{TP+FN}$

### **F1-Score**
Der F1-Score kombiniert die Metriken Recall und Precision und berechnet sich aus dem Harmonischen Mittel beider Metriken, welches mit folgender Formel berechnet werden kann. [5, p.4]

$F1-Score =  \frac{2 \cdot Precision\ \cdot Recall}{Precision\ +\ Recall}$

Diese Metrik wird hauptsächlich bei unausgeglichenen Datensätzen verwendet, wo sie besser geeignet ist als die Genauigkeit. Unausgewogene Datensätze liegen vor, wenn eine Klasse deutlich häufiger vorkommt als die andere. Angenommen, Klasse A kommt zu 98% und Klasse B zu 2% vor. Würde man nun pauschal sagen, dass ein Datenpunkt immer zur Klasse A gehört, läge man zu 98% richtig, was natürlich sehr gut aussieht, aber nur auf die Verteilung der Klassen zurückzuführen ist. [4, Ch. 5 Evaluierung und Verbesserung von Modellen: Metriken zu binären Klassifikatoren]

### **ROC-Kurve**
Die Fläche unterhalb der ROC-Kurve gilt als eine der bekanntesten Metriken zur Evaluation von modellen, da damit die Gesamtleistung des Modells abgebildet werden kann. Für einen binären Klassifikator lässt sich die Fläche mittels folgender Formel bestimmen.
$AUC \ = \ \frac{S_p-\frac{n_p \cdot (n_n+1)}{2}}{n_p \cdot \ n_n}$

Dabei steht $S_p$ für die Summer aller positiv klassifizierter Datenpunkte und $n_n$ für die Anzahl negativ klassifizierter Datenpunkte und äquivalent $n_p$ für die Anzahl aller positiv klassifizierter Datenpunkte.

[5, p.5]

Dementsprechend wird der Recall gegen den Fallout aufgetragen. Die ROC-Kurve sollte immer so weit wie möglich links-oben liegen. Ein geeignetes Maß zum Vergleich von ROC-Kurven ist die Area under the curve (AUC). Sie misst, wie der Name schon sagt, die Fläche unter der ROC-Kurve. Eine Fläche von 1 würde also einem perfekten Klassifikator entsprechen. [1, Ch. 3 Classification: Performance Measues]

## **5.3 Gaussian Naive Bayes**
Der Gaussian-Naive-Bayes-Algorithmus basiert auf dem Bayes'schen Satz und der Wahrscheinlichkeitstheorie [1]. Es wird angenommen, dass die einzelnen Merkmale des Datensatzes voneinander unabhängig sind [14, p.119]. Da dies jedoch selten in der Realität auftritt, wird eine Gauß'sche Verteilung der Merkmale im Bezug auf die Klassenzugehörigkeit angenommen [15, p.209]. Für diesen Algorithmus sind vor allem die bedingten Wahrscheinlichkeiten von Bedeutung. Darunter versteht man die Wahrscheinlichkeit des Auftretens einer bestimmten Klasse unter der Bedingung eines bestimmten Wertes eines Merkmals [14, p.119]. Unter der vorherigen Annahme der Gauß'schen Verteilung wird diese mit folgender Formel berechnet: $p(x_i|c)=\frac{1}{\sqrt{2 \pi \sigma ^2}}\cdot exp^{(-\frac{(x_i-\mu_{c, i})^2}{2\sigma_{c, i}^2})}$ [15, p.209]. Die Wahrscheinlichkeit einer Klasse ergibt sich aus der Multiplikation aller bedingten Einzelwahrscheinlichkeiten [14, p.119]. Die zugehörige Formel lautet $p(x_1,...,x_n|c)=\prod_{1}^n\ p(x_i|c)$ [15, p.209]. Letztendlich werden bei der Prognose alle Klassenwahrscheinlichkeiten berechnet und die mit der höchsten Wahrscheinlichkeit ausgewählt. [14, p.119]

## **5.4 K-nearest neighbour (KNN)**

Der KNN-Algorithmus (K-Nextest-Neighbor) basiert auf der Idee, dass die nächstgelegenen Werte zu einem Zielpunkt x Informationen über dessen Klasse liefern können. Bei diesem Algorithmus wird der Abstand aller Punkte im Datensatz zum Zielpunkt berechnet. Die Distanz wird in den häufigsten Fällen durch die euklidische Distanz gemessen. Nur die k Punkte mit dem geringsten Abstand zum Zielpunkt werden berücksichtigt. Innerhalb der k Punkte wird dann gezählt, wie oft jede Klasse vertreten ist und der Zielpunkt x wird der Klasse zugeordnet, die in den k Punkten am häufigsten vertreten ist.  Die Wahl von k spielt bei diesem Algorithmus eine besonders wichtige Rolle, da sie die Anzahl der Nachbarn bestimmt, die berücksichtigt werden. Im Allgemeinen kann man sagen, dass größere Werte von k besser sind, da sie präziser sind und das Rauschen verbessern können. Bei einem k-Wert von 1 wählt der Algorithmus immer die Klasse des nächsten Nachbarn von k. Ein k-Wert von 1 sollte daher nicht verwendet werden, da das Modell dann overfitten wird. Aus diesem Grund wird empfohlen, einen höheren Wert für k zu verwenden. Der Vorteil des KNN ist seine Einfachheit und die kurze Laufzeit.

Bei der Wahl der Hyperparameter haben wir uns für ungerade k-Werte entschieden, da dies die Entscheidungsfindung vereinfacht, wenn zwei Klassen in den k Punkten gleich verteilt sind. Außerdem haben wir uns entschieden, mit einem k-Wert von 5 zu beginnen, um ein Overfitting zu vermeiden. Da wir eine lange Laufzeit der GridSearch verhindern wollten, haben wir nicht alle möglichen Kombinationen der Hyperparameter in der GridSearch betrachtet. Beispielsweise haben wir uns im Vorfeld für die euklidische Distanzmetrik entschieden, da diese in der Literatur am häufigsten erwähnt wird. Alternativ könnte die Distanz auch mit der Manhattan-Metrik gemessen werden. Mit dem Parameter Weight kann zusätzlich angegeben werden, ob die Nachbarn einheitlich gewichtet werden, d.h. ob jeder Nachbar bei der Entscheidung gleich gewichtet wird, oder ob die Nachbarn nach der Distanz gewichtet werden und die näheren Nachbarn bei der Entscheidung für die Klasse eine größere Rolle spielen.

## **5.6 Stochastic Gradient Descent (SGD) Classifier**

Bei der Gradientenabsenkung sind in der Regel viele Stufen erforderlich, um ein lokales Minimum zu erreichen. Zur Beschleunigung dieses Prozesses gibt es in der Praxis eine Reihe von Ansätzen. Eines dieser Verfahren ist der SGD. Der SGD implementiert eine regularisierte Methode der Gradientenabsenkung. Wenn n groß ist, können wir, indem wir über alle n Beobachtungen summieren, jedes Mal, wenn wir einen Gradientenschritt berechnen, einen kleinen Bruchteil oder eine Minibatch verwenden. Anstatt den gesamten Datensatz zu nutzen, berechnet der SGD den Fehler und den Gradienten für eine Teilmenge der Datenpunkte und aktualisiert den Parameter entsprechend. Dies führt zu einer schnelleren Gradientenberechnung.

## **5.7. Random forest**

Random Forest ist ein Regressionsalgorithmus zur Vorhersage von Werten oder Klassen, der auf dem bestehenden Konzept der Entscheidungsbäume aufbaut. In einem Random Forest werden k Entscheidungsbäume trainiert und anschließend der Durchschnitt aller Ergebnisse berechnet. Die einzelnen Entscheidungsbäume werden auf unterschiedlichen Daten des gesamten Datensatzes trainiert, um Korrelationen innerhalb der Bäume zu vermeiden. Außerdem kann pro Entscheidungsbaum nur das beste Merkmal statt einer Kombination mehrer Entscheidungskriterien genutzt werden, um ebenso die Korrelation zwischen den Bäumen zu senken, was wiederum den Fehler des Modells reduziert, jedoch gleichzeitig auch die Stärke des Baums schwächt.  Zur Auswahl der verschiedenen Trainingsdaten wird das so genannte Bagging verwendet. Dabei wird aus den Daten ein Trainingsdatensatz gebildet, der die gleiche Größe wie der Ursprungsdatensatz hat, wobei einige Datenpunkte mehrfach und andere gar nicht im Trainingsdatensatz berücksichtigt werden. Dadurch wird der Trainingsdatensatz robuster gegenüber Schwankungen im Ursprungsdatensatz und die Vorhersagequalität steigt. Die im Bagging-Prozess nicht berücksichtigten Datenpunkte dienen schließlich zur Validierung des k-ten Modells, so dass nicht auf separate Validierungsdaten zurückgegriffen werden muss. Ein Generalisierungsfehler wird berechnet. Dieser konvergiert mit zunehmender Anzahl der Entscheidungsbäume, wodurch eine Überanpassung des Modells an die Daten vermieden wird. Neben den vorhergesagten Klassen bzw. Werten ermittelt der Random Forest auch den Einfluss der Merkmale des Datensatzes auf das Ergebnis. Dazu wird in jedem Durchlauf ein Merkmal im Datensatz ausgetauscht, während die anderen konstant im Datensatz bleiben. Die Genauigkeit des neuen Modells wird dann mit der des vorherigen Modells verglichen. [12, p.806]

## **5.8 Support Vector Machine - SVM**
Die SVM ist ein maschinelles Lernmodell, das lineare und nichtlineare Klassifikation und Regression durchführen kann. Die SVM versucht, eine Entscheidungsgrenze zu finden, die zwei Klassen trennt, aber immer versucht, den größtmöglichen Abstand zum nächsten Datenpunkt der Klasse zu haben. Man kann sich die SVM auch als die breiteste Straße zwischen zwei parallelen Linien vorstellen, die zwischen den beiden Klassen verläuft. Eine Variante der SVM ist die "hard margin". Hier wird angenommen, dass alle Instanzen der Klasse nicht auf dieser Straße liegen dürfen und dass sie auf der richtigen Seite des Klassifikators liegen. Die hard margin schränkt jedoch das Modell und die Breite der Straße stark ein. Daher gibt es die soft margin Klassifikatoren. Bei dieser Methode wird versucht, die Straße so breit wie möglich zu halten, aber auch die Grenzwertverletzungen klein zu halten. Wenn die Daten nicht linear verteilt sind, können polynomiale Merkmale hinzugefügt werden. Wenn das Modell sehr komplex ist, muss man beachten, dass sehr viele Merkmale erstellt werden und das Modell sehr langsam wird.  

Die SVM eignet sich besonders gut für unseren Schlaganfalldatensatz, da sie besonders gut auf kleinen und mittleren Datensätzen funktioniert. Der Regularisierungs-Hyperparameter c hilft uns zu bestimmen, wie groß das Grid sein soll und wie viele Grenzverletzungen wir zulassen wollen. Dabei ist zu beachten, dass ein zu kleiner c-Wert Underfitting und ein zu großer c-Wert Overfitting unterstützt. Zusätzlich haben wir uns entschieden, die SVM auf verschiedenen PolynomialDegressoren zu testen, um auch bei Nichtlinearität ein geeignetes Modell zu finden. Mit dem Parameter coef0 wollen wir zusätzlich kontrollieren, wie das Modell von höheren oder niedrigeren Graden beeinflusst wird. Aus Laufzeitgründen Testen wir in dieser Gridsearch nur den Polynominal Degree und setzen c und coef0 auf 1 fest.

## **5.9 Neuronales Netzwerk**

Künstliche Neuronale Netzwerke funktionieren ähnlich wie biologische Nervensysteme des menschlichen Körpers, wie beispielsweise dem Gehirn. Mehrere miteinander verbundene Rechenknoten, die äquivalent wie im Gehirn als Neuronen bezeichnet werden, arbeiten zusammen, um die Ausgabe durch die Eingaben zu optimieren. Üblicherweise wird ein mehrdimensionaler Vektor als Eingabe in die Eingabeschicht (input layer) geladen, die ihn anschließend an die verborgenen Schichten (hidden layers) weiterleitet. Beim Training treffen diese Entscheidungen basierend aus den Ergebnissen der vorherigen Schicht und wägen ab, ob eine stochastische Veränderung die Ausgabe verbessert oder verschlechtert. # **6. Gesamtauswertung aller Algorithmen**
Zur Bewertung des Modells werden im Wesentlichen drei Metriken herangezogen. Besonderes Augenmerk legen wir dabei auf die Sensitivität (Recall). Diese misst, wie bereits beschrieben, den Anteil der richtig als positiv klassifizierten Fälle an allen als positiv klassifizierten Fällen. Bezogen auf Schlaganfälle bedeutet dies, dass diese Metrik eine Aussage darüber trifft, wie gut das Modell in der Lage ist, echte Schlaganfälle zu erkennen. Ein hoher Wert der Sensitivität bedeutet also, dass das Modell die meisten Schlaganfälle erkennt. Ein niedriger Wert der Sensitivität bedeutet hingegen, dass das Modell viele mögliche Schlaganfälle nicht erkennt. Zusätzlich zum Recall, aber nicht so stark gewichtet, betrachten wir die Precision. Diese gibt wie beschrieben an, wie viele der als positiv klassifizierten Fälle tatsächlich positiv sind. Für Schlaganfälle bedeutet dies, dass die Präzision angibt, wie viele der vom Modell als Schlaganfall identifizierten Fälle tatsächlich Schlaganfälle sind. Die Genauigkeit ist wichtig, um die Anzahl der Fehlalarme zu minimieren. Wir halten die Sensitivität für wichtiger, da es aus unserer Sicht von Bedeutung ist, dass möglichst viele echte Schlaganfälle erkannt werden. Dafür würden wir auch eine geringere Genauigkeit in Kauf nehmen und einen Schlaganfall vorhersagen, obwohl keiner vorliegt. Um die verschiedenen Modelle dahingehend zu bewerten, haben wir das Modell mit den bedien Metriken aufgetragen. Eine weitere Metrik, die wir zur Bewertung der Modelle herangezogen haben, ist der F1-Score. Dieser ist das harmonische Mittel aus Präzision und Sensitivität. Ein hoher F1-Score bedeutet also für das Schlaganfallmodell, dass es die meisten tatsächlichen Schlaganfälle erkennt und gleichzeitig die Anzahl der falsch positiven Vorhersagen minimiert.
